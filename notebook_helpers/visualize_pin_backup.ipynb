```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

```python
df = pd.read_csv("/home/robbin/MuMDIA_gh/MuMDIA/outfile.pin", sep="\t")
```

```python
df_targets = df[df["Label"] == 1.0]
df_decoys = df[df["Label"] == -1.0]
```

```python
selected_peptides = ["LSDYGVQLR", "ENYLIDNLDR", "LIDQATAEIVETAK"]
for c in df.columns:
    if pd.api.types.is_numeric_dtype(df[c]):
        plt.hist(
            df_targets[c],
            alpha=0.5
        )
        plt.hist(
            df_decoys[c], 
            alpha=0.5
        )
        plt.title(c)

        for pep in selected_peptides:
            plt.scatter(
                df[df["Peptide"] == pep][c], [10] * len(df[df["Peptide"] == pep][c])
            )
        plt.show()
```

```python
plt.scatter(
    df_targets["peptide_q_mean"],
    df_targets["top_correlation_matrix_psm_ids_3"],
    alpha=0.1,
    s=1,
)
plt.scatter(
    df_decoys["peptide_q_mean"],
    df_decoys["top_correlation_matrix_psm_ids_3"],
    alpha=0.1,
    s=1,
)
```
